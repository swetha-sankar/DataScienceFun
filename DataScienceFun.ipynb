## Box Plots

A boxplot (also known as a 'box and whisker plot') is a good visualization of the **five number summary** of a dataset. 
- Minumum
- Quartile 1 (Q1)
- Median 
- Quartile 3 (Q3)
- Maximum

## How to create a boxplot from a Pandas DataFrame
The only argument needed to plot by column is the column name as a string. 

Ex: ```DataFrame.boxplot("colname")```


Default values:
```DataFrame.boxplot(column=None, by=None, ax=None, fontsize=None, rot=0, grid=True, figsize=None, layout=None, return_type=None, backend=None, **kwargs)```

or (with matplotlib.pyplot imported as plt):

```DataFrame.plot(kind = 'box')```
<br />
```plt.show()```

## Create a boxplot for the following dataset

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.randn(20, 3),
                  columns=['Col1', 'Col2', 'Col3'])

df.plot(kind = 'box')
plt.show()

#or:

df.boxplot(column = ["Col1", "Col2", "Col3"])
````

````python
#EXAMPLE
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd
energydf = pd.read_csv("https://corgis-edu.github.io/corgis/datasets/csv/energy/energy.csv")
energydf 
````

````python
myhistdf = pd.DataFrame(energydf)
myhistdf.hist(column="Consumption.Commercial.Liquefied Petroleum Gases")
plt.title("Title")
plt.xlabel("Frequency")
plt.ylabel("Commercial Consumption of Liquefied Petroleum Gases")
plt.show()
````
